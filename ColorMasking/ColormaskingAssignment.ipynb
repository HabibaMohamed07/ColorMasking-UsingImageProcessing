```python
import cv2
import numpy as np
from google.colab.patches import  cv2_imshow
from google.colab import drive
drive.mount('/content/gdrive')

img = cv2.imread("/content/gdrive/MyDrive/color_pencils.jpeg", cv2.IMREAD_UNCHANGED)

hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

mask = cv2.inRange(hsv , (25,150,20), (30,255,255)) #el range da aktar range precise ashan ageeb l loonl asfar lwhdo
maskk = cv2.morphologyEx(mask, cv2.MORPH_OPEN, np.ones((3,3), np.uint8))

output = cv2.bitwise_and(img, img, mask= maskk)


cv2_imshow(img)
cv2_imshow(maskk)
cv2_imshow(output)



#closing bytla3 hagat sah bs fe noise, so opening ahsan.

# maskk2 = cv2.morphologyEx(mask, cv2.MORPH_CLOSE, np.ones((3,3), np.uint8))
# output2 = cv2.bitwise_and(img, img, mask= maskk2)
# cv2_imshow(maskk2)
# cv2_imshow(output2)
```

```python
import cv2
import numpy as np


img = cv2.imread("/content/gdrive/MyDrive/rainbow.png", cv2.IMREAD_UNCHANGED)

hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

mask = cv2.inRange(hsv , (25,150,20), (30,255,255))
mask = cv2.morphologyEx(mask, cv2.MORPH_OPEN, np.ones((3,3), np.uint8))
output = cv2.bitwise_and(img, img, mask= mask)


cv2_imshow(img)
cv2_imshow( mask)
cv2_imshow( output)
```
